```python
import cv2
import matplotlib.pyplot as plt
import os

# load image
img_path = "train/images/" + os.listdir("train/images")[0]
label_path = "train/labels/" + os.listdir("train/labels")[0]

img = cv2.imread(img_path)
h, w, _ = img.shape

# read labels
with open(label_path) as f:
    lines = f.readlines()

for line in lines:
    cls, x, y, bw, bh = map(float, line.split())

    # convert YOLO format to pixel coordinates
    x1 = int((x - bw/2) * w)
    y1 = int((y - bh/2) * h)
    x2 = int((x + bw/2) * w)
    y2 = int((y + bh/2) * h)

    cv2.rectangle(img, (x1, y1), (x2, y2), (0,255,0), 2)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
```

```python
from ultralytics import YOLO

model = YOLO("runs/detect/train3/weights/best.pt")

metrics = model.val()
```

```python
print("Model Performance Report")
print("------------------------")

print("Precision:", round(metrics.box.p.mean()*100,2), "%")
print("Recall:", round(metrics.box.r.mean()*100,2), "%")
print("mAP50:", round(metrics.box.map50*100,2), "%")
print("mAP50-95:", round(metrics.box.map*100,2), "%")
```
